```python
import requests
from bs4 import BeautifulSoup 
from tqdm import tqdm
import re
import pandas as pd
import time
import shutil
import os
from os import listdir
from os.path import isfile, join
```

```python
path = './scrapped'
files = [f for f in listdir(path) if isfile(join(path, f))]
files
```

```python
pattern = re.compile('(part|chapter|article|section|sub-section|paragraph|subparagraph|item|\
                        general provisions|common provisions|supplementary provisions|penalty provisions|\
                        addendum (addenda)|enforcement date|transitional measures|attached table| form|\
                        제.*[편장조절관조항호목])')
target = []
not_exist = []

for file in tqdm(files):
    try:
        tmp = pd.read_excel('./scrapped/{}'.format(file), index_col=0)
        TF_01 = re.search(pattern, tmp.iloc[2].Eng_Rough_Ver.lower()) == None
        TF_02 = re.search(pattern, tmp.iloc[2].Kor_Rough_Ver.lower()) == None
        if TF_01 == True or TF_02 == True : 
            target.append('{}'.format(file))

    except FileNotFoundError :
        not_exist.append('{}'.format(file))
        continue
        
    except AttributeError :
        target.append('{}'.format(file))
        continue
    except Exception as e :
        print(e)
        target.append('{}'.format(file))
        continue
        
print('target >>> ', target)
print('not_exist >>> ', not_exist)
```
