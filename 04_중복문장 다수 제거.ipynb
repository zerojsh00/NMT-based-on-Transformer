```python
import pandas as pd
from collections import OrderedDict
from tqdm import tqdm
```

```python
df = pd.read_excel('./rawdata_190723.xlsx', index_col=0)
df.head()
```

```python
print(len(df))
unique = OrderedDict()

for i, row in tqdm(df.iterrows()):
    if row['ENG'] not in unique.values() :
        unique[i] = (row['ENG'])
        
print(len(unique))
```

```python
len(df)
```

```python
unique.keys()
```

```python
unique_idx = list(unique.keys())
unique_idx
```

```python
df = df.iloc[unique_idx]
len(df)
```
