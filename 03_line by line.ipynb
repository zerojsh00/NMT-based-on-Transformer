### 로우별로 다시 처리 될 수 있도록 ..

```python
revision_pattern = re.compile('[<\[].*(amended|act no|date|decree no|개정|신설).*[>\]]')

completed = []
df_line = pd.DataFrame()

for file in tqdm(file_list) :
    # 파일 단위 리스트
    ENG = []
    KOR = []
    
    # 하나의 로우 단위로 처리
    for i, row in df[df.File_Name == '{}'.format(file)].iterrows() :
        eng_split = [] 
        kor_split = []
        
        # 각 로우의 문장들이 담긴 리스트
        eng_split += preprocessor(row['Eng_Rough_Ver']).split('\n')
        kor_split += preprocessor(row['Kor_Rough_Ver']).split('\n')

        # ''를 제거한 리스트
        eng_split = [line for line in eng_split if line !='']
        kor_split = [line for line in kor_split if line !='']


        # 리스트 단위로 직전 문장의 반복을 제거
        temp_list = []
        temp = ''
        for line in eng_split:
            if temp == line:
                pass
            else:
                temp_list.append(line)
            temp = line
        eng_split = temp_list

        temp_list = []
        temp = ''
        for line in kor_split:
            if temp == line:
                pass
            else:
                temp_list.append(line)
            temp = line
        kor_split = temp_list
        
    
        # 로우 단위에서 부칙 및 개정관련 <pad> 처리
        for j in range(min(len(eng_split), len(kor_split))) :
            TF_01 = re.match(addenda_pattern, eng_split[j].lower().strip()) != None
            TF_02 = re.match(addenda_pattern, kor_split[j].lower().strip()) != None

            if TF_01 == True and TF_02 == False :
                kor_split.insert(j, '<pad_부칙관련>')
                continue
            elif TF_01 == False and TF_02 == True :
                eng_split.insert(j, '<pad_부칙관련>')
                continue

            TF_01 = re.search(revision_pattern, eng_split[j].lower().strip()) != None
            TF_02 = re.search(revision_pattern, kor_split[j].lower().strip()) != None

            if TF_01 == True and TF_02 == False :
                kor_split.insert(j, '<pad_개정관련>')
                continue

            elif TF_01 == False and TF_02 == True :
                eng_split.insert(j, '<pad_개정관련>')
                continue
                
        # 로우 단위에서 최종 길이가 같으면 파일 단위 리스트에 추가
        if len(eng_split) == len(kor_split) :
            ENG += eng_split
            KOR += kor_split

    # 파일 단위에서 데이터 프레임 생성
    a = pd.DataFrame(ENG, columns=['ENG'])
    b = pd.DataFrame(KOR, columns=['KOR'])
    c = [file] * (max(len(a), len(b)))
    
    data = pd.merge(a, b, how='outer', left_index=True, right_index=True)
    data['File_Name'] = c
    data.fillna('', inplace = True)

    
    # 부칙이 아닌 곳에서 '생략' 뜨면 그 밑은 지워버리기
    for idx, data_row in data.iterrows() :
        if ('omitted' or 'deleted') in data_row['ENG'].lower().strip() :
            data = data.iloc[:idx]
            break
        elif ('삭제') in data_row['KOR'].lower().strip() :
            data = data.iloc[:idx]
            break
            

    completed.append(file)
    df_line = df_line.append(data).reset_index(drop=True)

df_line.fillna('', inplace = True)
        
```
